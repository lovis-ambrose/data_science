```python
# creating na values
df=pd.read_csv("csv_files/stock_data.csv", header=None, names=["tickers","eps","revenue","price","people"],na_values=["not available","n.a."])
df
```

```python
# great but we cant have negative revenue, to fix that, use a dictionary instead of a list. a list will convert all the negatives to NaN
df=pd.read_csv("csv_files/stock_data.csv", header=None, names=["tickers","eps","revenue","price","people"],na_values={
    "eps": ["not available","n.a."],
    "revenue": ["not available","n.a.",-1],
    "price": ["not available","n.a."],
    "people": ["not available","n.a."]
})
df
```

```python
# convert dataframe to excel
# write a function to replace na with a persons name
# first read the file
def convert_people_cell(cell):
    if cell == "n.a.":
        return "sam walton"
    return cell
# retur NaN where there is not available
def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell
# also we cant have a negative revenue
def convert_revenue_cell(cell):
    if cell == -1:
        return None
    return cell


df=pd.read_excel("csv_files/stock_data.xlsx",header=None, names=["tickers","eps","revenue","price","people"], converters={
    "people": convert_people_cell,
    "eps": convert_eps_cell,
    "revenue": convert_revenue_cell
})
df
```

```python
# now convert to excel
# set index to None to remove indices in the xlsx file
df.to_excel("csv_files/new.xlsx",index=None)
```

Handling missing data in pandas

```python
# read the data frame
df=pd.read_csv("csv_files/weather_data.csv")
df
```

```python
# convert day from str to date
df=pd.read_csv("csv_files/weather_data.csv", parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
# set day as the index
df.set_index("day", inplace=True)
df
```

```python
# replace NaN with zeros. I dont wanna change my original df
new_df = df.fillna(0)
new_df
```

```python
# but event cant have zeros, use a dicct to cater for the valus
new_df=df.fillna({
    "temperature": 0,
    "windspeed": 0,
    "event": "no event"
})
new_df
```

```python
# fill the missing gaps with forward fill ie ffill or backward fill ie bfill
new_df=df.ffill()
new_df
```

```python
# fill the missing gaps with forward fill ie ffill or backward fill ie bfill
# you can as well set the limit arg inside the ffill(), ie choose the no of consecutive gaps to fill 
# eg limit = 1 would fill only one forward gap(copy once) leaving the proceeding as NaN
new_df=df.ffill(limit=1)
new_df
```

```python
# use interpolation for better guesses than just fillna
# specifying method to time comes up with a better guess. method is linear by default
new_df=df.interpolate(method="time")
new_df
```

```python
# drop rows with na values
new_df=df.dropna()
new_df
```

```python
# drop the rows without any single value
# drops the row thats entirely empty ie 2017-01-09
new_df=df.dropna(how="all")
new_df
```

```python
# keep the row that has atleast 1 valid value and drop the rest
new_df=df.dropna(thresh=1)
new_df
```

```python
# fill the missing dates eg from 1st jan to 11th jan
dt=pd.date_range("01-01-2017","01-11-2017")
idx=pd.DatetimeIndex(dt)
new_df=df.reindex(idx).interpolate()
new_df
```

Meging data frames

```python
# first data frame
df1 =pd.DataFrame({
    "city":["new york","chikago","atlanta"],
    "temperature":[66,65,75],
})
df1
```

```python
# second data frame
df2=pd.DataFrame({
    "city":["new york","chikago","atlanta"],
    "humidity":[25,30,29]
})
df2
```

```python
# merge the data frames
df3 = pd.merge(df1, df2, on="city")
df3
```

Consider a case where all the columns are not the same. Here do an outer join. ie how="outer"

```python
# first data frame
dff1 =pd.DataFrame({
    "city":["new york","chikago","missouri"],
    "temperature":[66,65,75],
})
dff1
```

```python
# second data frame
dff2=pd.DataFrame({
    "city":["new york","new jersey","missouri","toronto"],
    "humidity":[25,30,29,26]
})
dff2
```

```python
# how is set to inner by default. it is an equivalent of intersection
# use how="outer" in this case. ==> union
# see also right join, left join....
# compare this to a set theory
newdf=pd.merge(dff1,dff2, on="city", how="outer")
newdf
```

```python
# dealing with data frames that have many columns in common
# joining two dataframes
ddf1=pd.DataFrame({
    "city":["new york","chikago","atlanta","delhi"],
    "temperature":[21,14,35,24],
    "humidity":[66,65,75,70],
})
ddf1
```

```python
ddf2=pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[25,30,29],
    "humidity":[68,60,78]
})
ddf2
```

```python
# explicitly specify from which table the data came with suffixes function
df2=pd.merge(ddf1, ddf2, on="city", how="outer",suffixes=["_left","_right"])
df2
```

# pivot table allows you to summarize and aggregate data inside the data frame

```python
# lets have two values of the same day for ach city eg morning and evening temperature $ humidity
df=pd.read_csv("csv_files/weather2.csv")
df
```

```python
# note that index represents rows
# pivote table will find the average temp and humidity per day
# we can say the input data frane is aggregated
pd.pivot_table(df, index="city", columns="date")
```

```python
# note that index represents rows
# pivote table will find the average temp and humidity per day
# we can say the input data frane is aggregated
# we can also specify the aggregate function eg sum, its set to average by default
# set margins to true to sum temp and humidity for both cities per day for both index and columns
# change margins_name to your liking, its set to all by default
pd.pivot_table(df, index="city", columns="date", aggfunc="sum",margins=True, margins_name="Total per day")
```

# Aggregating based on date frequency

```python
# consider the data frame below
# notice that it has data for two months, may and dec.
df=pd.read_csv("csv_files/weather3.csv")
df
```

```python
# Now we want eg to know the monthly averahge temp and humidity for each month
# we use grouper function
# first we need to convert date (str) to date(timestamp)
df["date"]=pd.to_datetime(df["date"])
df
```

```python
# confirm that its now date (timestamp)
type(df["date"][0])
```

```python
# Now we want eg to know the monthly average temp and humidity for each month
# we use grouper function
df.pivot_table(index=pd.Grouper(freq="M", key="date"), columns="city")
```

# Stack, unstack
Used to format

```python
# Read the xlsx file with MultiIndex header, skipping the first row
df5 = pd.read_excel("csv_files/stocks.xlsx", header=[0, 1], sheet_name="stock_price", parse_dates=[0])
df5
```

```python
# stack the dataframe
df_stacked=df5.stack()
df_stacked
```

```python
# stack the dataframe
#  define the Level(s) to stack from the column axis onto the index axis, defined as one index or label, or a list of indices or labels.
# level =0 means moving first part of the column index to the rows
# level =1 means moving 2nd part of the column index to the rows
df_stacked=df5.stack(level=1)
df_stacked
```

```python
# unstack df to original
df_stacked.unstack()
```

# Contingency table / cross tabulation
A table in matrix form that displays the (multivariate) frequency distribution of the variables

```python
# read the file
df7=pd.read_excel("csv_files/crosstab.xlsx")
df7
```

```python
# contingency table/crosstab
# The first argument specifies variable you want to put at a row level (x-axis)
# second argument specifies the columns (y-axis)
pd.crosstab(df7.Nationality, df7.Handedness)
```

```python
# i want to know how many females and males per country
pd.crosstab(df7.Nationality, df7.Sex)
```

```python
# know how many people per country
pd.crosstab(df7.Nationality, df7.Nationality)
```

```python
# i want to kow the freq distribution of handedness per sex
# To know the total number, use margins arg
pd.crosstab(df7.Sex, df7.Handedness, margins=True, margins_name="Total")
```

```python
# know how many are from which country and their freq distribution
pd.crosstab(df7.Sex,[df7.Handedness, df7.Nationality], margins=True, margins_name="Total")
```

```python
# know how many are from which country and their freq distribution
pd.crosstab([df7.Sex, df7.Nationality],df7.Handedness, margins=True, margins_name="Total")
```

```python
pd.crosstab([df7.Nationality, df7.Sex],df7.Handedness, margins=True, margins_name="Total")
```

```python
# i want to get the percentage of the individual vales
pd.crosstab(df7.Sex, df7.Handedness, margins="True")
```

```python
# i want to get the percentage of the individual vales
pd.crosstab(df7.Sex, df7.Handedness, normalize="index")
```

```python
# i want to know the average age of eg females who are left handed and also for males
# to do this supply values (Age column) and set aggfunc to numpy functionaverage
# we need numpy to compute the average age for the handedness
pd.crosstab(df7.Sex, df7.Handedness, values=df7.Age, aggfunc=np.average)
```
