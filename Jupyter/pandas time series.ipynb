# Handling time holidays in time series analysis and how to create a custom holiday calender and generate datetimeindex out of it

```python
# read the csv file. Notice that we have no data on 4th of july 2017 coz its a public holiday for US and stocks dont trade
# 
th_df = pd.read_csv("csv_files/aapl/july16_17.csv")
th_df.head(30)
```

```python
# now import import UsFederalHolidayCalender
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay
```

```python
# create a custom US business day
usb= CustomBusinessDay(calendar=USFederalHolidayCalendar())
usb
```

```python
# now use the custom calendar as frequence
# notice that 4th jul is excluded now coz its a holiday
ndr= pd.date_range(start="7/1/2017", end="7/21/2017",freq=usb)
ndr
```

```python
# we can now set our new date range to our dataframe
# This will exclude weekends and holidays
df_miss.set_index(ndr, inplace=True)
df_miss
```

# customizing the holiday calendar

```python
# copy the abstract class for pandas calendar from github
# available at:   https://github.com/pandas-dev/pandas/blob/main/pandas/tseries/holiday.py

# import the necessary classes
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday


class MyBirthDayCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday("Mutungi's Birthday", month=7, day=30, observance=nearest_workday),
    ]

# now custom your own calendar
# execute it to create an instance of MyBirthDayCalendar class
# when you set observance=nearest_workday, then 29th will also be excluded
# holiday cal
# nearest_workday == move sat to fri and sun to mon
# see various values you can supply as observance
mycal = CustomBusinessDay(calendar=MyBirthDayCalendar())
mycal
```

```python
# now supply it in the pandas daterange function
# from the output, we can see that 2nd jul and 30th jul are excluded from the data coz they are holidays for Nam and Mutungi
pd.date_range(start="7/1/2020", end="10/30/2020", freq=mycal)
```

want to cater for countries where friday and saturday are holidays/weekends eg muslim countries like Egypt

```python
# we can do this by customizing the CustomBusinessDay's weekmask function
# lets say 29th oct is a holiday, just include it also
egypt = CustomBusinessDay(weekmask="Sun Mon Tue Wed Thu", holidays=["2023-10-29"])
egypt
```

```python
# now pass the created custom business days to the dataframe
# the fridays and satardays are now treated as weekend
# notice that days like 6,7, 13,14,20,21,27,28 are excluded and treated as weekends
# also 29th is excluded as a holiday as explicitly made
pd.date_range(start="10/1/2023", end="10/30/2023", freq=egypt)
```

# shifting and lagging
Shifting can be used to calculate percentage change

```python
# read fb csv file
fb=pd.read_csv("csv_files/fb.csv")
fb
```

```python
fb.set_index("Date", inplace=True)
fb
```

```python
# now shift the prices by one day
# right shift
# will shift the price of 25th to 26th
fb.shift(1)
```

```python
# left shifting
# this will shift the price of 16th to 15th
fb.shift(-1)
```

```python
# lets use shift to calculate price change in one day
# create a column for the previous day
fb["Price for previous day"]= fb["Price"].shift(1)
fb
```

```python
# now calculate the price change
fb["Price change in 1 day"] = fb["Price"] - fb["Price for previous day"]
fb
```

```python
# calculate the five day percentage return
fb["5 day % return"] = (fb["Price"] - fb["Price"].shift(5))*100/fb["Price"].shift(5)
fb
```

# Handling time zones

```python
# read Microsoft data file
ms = pd.read_csv("csv_files/msdata.csv", header=1, index_col="Date Time", parse_dates=True)
ms
```

```python
# notice that the date below is a Naive datatime object
ms.index
```

```python
# There are two types of time zone
# 1. Naive (No timezone awareness). This is the default timezone
# 2. Timezone aware datetime
# change timezone
ms = ms.tz_localize(tz="US/Eastern")
ms
```

```python
# notice the diff in timezones eg '2017-08-17 09:00:00' and '2017-08-17 09:00:00-04:00'
# this means that US timezone is 4 hours behind the UTC/GMT
ms.index
```

```python
# lets convert the timezone now to time of Berlin, Germany
# notice that its 15:00:00+02:00 ie 15:00:00 in + 2 hrs ie Berlin is 2hrs ahead of GMT
ms = ms.tz_convert(tz="Europe/Berlin")
ms
```

```python
# Time strings
from pytz import all_timezones
all_timezones
```

```python
# converting to kampala timezone
# Notice that Kampala is 3hrs ahead of GMT
ms = ms.tz_convert("Africa/Kampala")
ms
```
