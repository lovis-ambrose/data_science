### multiclass classification
Eg predicting which party a person is going to vote for;

1. Democratc
2. Republican
3. Independent </br>
This is multiclass classification

```python
import matplotlib.pyplot as plt
import pandas as pd
# lets also import sklearn predefined datasets
from sklearn.datasets import load_digits
```

we want to recognize the hand written digits

```python
# load the trainng set
digits = load_digits()
```

```python
# now lets explore what this set contains
dir(digits)
```

We can see that it contains data ie the real data, images etc

```python
# lets see what is there
# we can see that it contains an 8*8 image represented by a 1D array 
digits.data[0]
```

```python
# lets print the corresponding image represented by the above array
plt.gray()
plt.matshow(digits.images[0])
```

```python
# lets print the first 5 samples
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
    plt.show()
```

```python
# we shall use target and data for training our model
digits.target[:5]
```

our dataset will be the digits.data, then target variable is digits.target

```python
# import the libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# lets now split our dataset intotraining and testing data
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
# length of training data ie 80%
len(X_train)
```

```python
# length of the test data ie 20%
len(y_test)
```

```python
# now we can train our dataset
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test)
```

```python
# lets predict some random value eg 67
plt.matshow(digits.images[77])
```

```python
# lets see the predicted number
digits.target[77]
```

```python
# lets predict 0:5
# compare the result with the previously predicted images
model.predict(digits.data[0:5])
```

### lets find out where the moddel is not doing well
We shall use confusion matrix method method

```python
# lets first get all the predicted values
y_predicted = model.predict(X_test)
# import confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
# in the confusion_matrix, suppy y_test ie the truth and the y_predicted
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
 # lets use seaborn to better visualise out results
import seaborn as sn
plt.figure(figsize=[10,7])
sn.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

Interpreting the chart.

1. the first number 32, truth was 0 and the model predicted 0 making it accurate
2. for 37, the image of 37 was 8 but the model predicted it as 1 (for two times)

Use sklearn datasets iris flower dataset to train your model using logistic regression. You need to figure out the acccuracy of your model and use that to predict different samples in your test dataset. In this iris dataset, there are 150 samples containing following features.

1. sepal length
2. sepal width
3. petal length
4. petal width


Using the above 4 features, you will classify the flower in one of the three categories

1. setosa
2. versicolour
3. virginica

```python
# import the dataset
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
# lets look at the contents of the dataset
dir(iris)
```

```python
# len of the dataset
len(iris.data)
```

```python
# lets look at our dataset names
iris.target_names
```

```python
# create a dataframe
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
# add the target column
df["target"] = iris.target
df
```

```python
# add target names
df["flower_name"] = df.target.apply(lambda x: iris.target_names[x])
df
```

```python
# lets import train_test_split and logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# lets split our dataset
X_train, X_test, y_train, y_test = train_test_split(df[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]], df.target, test_size=0.2)
```

```python
# lets look at the train and test dataset length
len(X_train), len(X_test)
```

```python
# train the model
# create an instance of the LogisticRegression class
iris_model = LogisticRegression(solver="liblinear")
iris_model.fit(X_train, y_train)
```

```python
# lets see the accuracy of the model
iris_model.score(X_test, y_test)
```

```python
# lets make some predictions
iris_model.predict(X_test)
```
